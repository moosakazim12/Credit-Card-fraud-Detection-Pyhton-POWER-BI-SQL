```python
sns.countplot(x='Class', data=df)

```

```python
sns.histplot(df['Amount'], bins=100)
```

```python
sns.boxplot(x='Class', y='Amount', data=df)
```

```python
df.groupby('Class')['Amount'].mean()
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
```

```python
sns.countplot(x='Hour', data=df[df['Class']==1])
```

```python
corr = df.corr()['Class'].sort_values(ascending=False)
corr
```

```python
plt.figure(figsize=(8,10))
sns.heatmap(df.corr()[['Class']], cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(df['Amount'])
```

```python
fraud_by_hour = df.groupby('Hour')['Class'].mean() * 100
fraud_by_hour
```

```python
fraud_by_hour.plot(kind='line', figsize=(10,5), title='Fraud Rate by Hour (%)')
```
